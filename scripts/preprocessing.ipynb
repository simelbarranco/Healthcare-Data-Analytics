### Exploring and Preprocessing using Python (Pandas & NumPy)

```python
import pandas as pd
import numpy as np

# Loading the dataset
hc_df = pd.read_csv('../data/raw/healthcare_dataset.csv')

# Quick look at the structure
print(hc_df.shape)
print(hc_df.dtypes)
hc_df.head(10)
```

```python
cleaned_df = hc_df.copy()
cleaned_df.head()
```

```python
# Removing duplicates
cleaned_df.drop_duplicates(inplace=True)

# Saving cleaned dataset
cleaned_df.to_csv("../data/cleaned/final_dataset.csv", index=False)

print("✅ Preprocessing complete! Cleaned data saved in 'data/cleaned/final_dataset.csv'.")
```
