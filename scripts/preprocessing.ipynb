```python
import pandas as pd

# Loading the dataset
hc_df = pd.read_csv('../data/raw/healthcare_dataset.csv')

# Quick look at the structure
print(hc_df.shape)
print(hc_df.dtypes)
hc_df.head(10)
```

```python
# Step 1: Creating a copy of the original dataset and using the copy moving forward
cleaned_df = hc_df.copy()
cleaned_df.head(3)
```

```python
# Step 2: Standardizing column names
cleaned_df.columns = cleaned_df.columns.str.strip().str.lower().str.replace(' ', '_')
cleaned_df.head(3)
```

```python
# Step 3: Checking for missing values, duplicates, and proper string casing
cleaned_df.drop_duplicates(inplace=True) 
cleaned_df['name'] = cleaned_df['name'].str.title() 
null_counts = cleaned_df.isnull().sum()
if null_counts.any():
    print("Null values found in the following columns:")
    print(null_counts[null_counts > 0])
else:
    print("No null values found in the dataset.")
```

```python
# Step 4: Fill missing age with median and removing trailing commas
if 'age' in cleaned_df.columns:
    cleaned_df['age'] = cleaned_df['age'].fillna(cleaned_df['age'].median())

if 'hospital' in cleaned_df.columns:
    cleaned_df['hospital'] = cleaned_df['hospital'].astype(str).str.rstrip(',')

cleaned_df.head(10)
```

```python
# Step 5: Rounding all numeric columns to two positions
def round_numeric_columns(df, decimals=2):
    numeric_columns = df.select_dtypes(include=['float', 'int']).columns
    df[numeric_columns] = df[numeric_columns].round(decimals)
    return df

round_numeric_columns(cleaned_df).head()
```

```python
# Step 6: Handling date columns conversion for further use
cleaned_df['date_of_admission'] = pd.to_datetime(cleaned_df['date_of_admission'], errors='coerce')
cleaned_df['discharge_date'] = pd.to_datetime(cleaned_df['discharge_date'], errors='coerce')
print(cleaned_df['date_of_admission'].head(3))
print(cleaned_df['discharge_date'].head(3))
```

```python
# Saving cleaned dataset
cleaned_df.to_csv("../data/cleaned/cleaned_dataset.csv", index=False)

print(" Preprocessing complete! Cleaned data saved in 'data/cleaned/cleaned_dataset.csv'.")
```
